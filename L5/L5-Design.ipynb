# 1. Aufgabenstellung
-----

Es sollte ein Scheduler entstehen, der verschiedene Mengen von gleichartigen Tieren verwaltet und diese an eine Menge von Futterplätzen verteilt. Es dürfen nur Tiere einer Spezies zu einem Zeitpunkt fressen. Dass eine Tierart nicht verhungert muss berücksichtigt werden; und auch die  Situation, dass es "zu viel" Futter gibt, z. B. mehr Futternäpfe als Tiere.

## Anforderungen
Die Anforderungen an den Scheduler waren:
- Faires Füttern der Tiere
- Nur eine Tierart darf zu einem Zeitpunkt essen
- Kein Verhungern
- Gute Auslastung der Resource
- Tiere dürfen während sie essen nicht unterbrochen werden(nonpreemptiv)
- Gute Skalierung bei kleinen und großen Mengen von Futter und Tieren

## 2.1 Schedulingverfahren
Die Strategie des Schedulers beruht hauptsächlich auf dem FIFO-Prinzip. Die Tiere, die zuerst in die Warteschlange gespeichert wurden, werden auch zuerst wieder daraus entnommen.
Um die Nutzung der Fressnäpfe zu optimieren, durchsucht der Scheduler die Tiertypen, der drei am längsten wartenden Tiere in der Warteliste. Wenn einer der Tiertypen, 
mit dem gerade essendem Tiertyp übereinstimmt, wird dieses Tier als nächstes zum essen ausgewählt.

Außerdem wird sichergestellt, dass kein Tier verhungert. Dazu gibt es einen Zähler, der beim Auswälen eines Tieres, in der Position zwei oder drei in der Warteschlange erhöht und bei Auswahl der Position 0 wieder zurückgesetzt wird. Erreicht der Zähler 3, 
wird gewartet, bis alle aktuelle essenden Tiere die Futterstellen verlassen. Das Tier an Positon 0(am längsten wartende Tier), dass jetzt drei mal übersprungen wurde, wird nun zum essen zugelassen.

### 2.4.1 Funktion animalfeeding(animal)
Diese Funktion beschreibt das Füttern der Tiere selbst. Sie wird von den Tierthreads ausgeführt.
* Zuerst wird auf die Erstellung aller anderen Tierthreads, durch eine Barrier, gewartet
* Eine Schleife wird so lange durchlaufen, wie oft das Tier, in den übergebenen Parametern, gefüttert werden soll:
    1. Das Tier wird in die Warteschlange, aller wartenden Tiere eingefügt(*s. 2.2 Globale Variablen:* **animals**)
    2. Das Tier wird in die Warteschlange, der spezifischen Tierliste eingefügt(*s. 2.2 Globale Variablen:* **cats/dogs/mice**)
    3. Es wird gewartet bis das Tier zum essen zugelassen wird. Umsetzung durch eine Condition Variable der jeweils spezifischen Tierliste(*s. 2.3* **cond**)
        * Auswahl erfolgt durch den Scheduler(s. 2.5.3), der mit notify das Tier freigibt
    4. Speichern der Wartezeit in tierspezifische Messdatenliste(*s. 2.2 Globale Variablen:* **me_cats/me_dogs/me_mice**)
    5. Nach der Freigabe wird das Tier wieder aus jeglichen Wartelisten gelöscht
    6. **dish_cv** wird gelockt um Essensbereich zu schützen
    7. Suche nach dem nächstfreien Essplatz in **dish_eating** (s. 2.2 Globale Variablen)
    8. Anzahl essender Tiere wird erhöht (s. 2.2 Globale Variablen: **num_eating_animals**)
    9. Scheduler wird benarichtigt, dass er weitermachen kann (s. 2.2 Globale Variablen: **sched_cv**)
    10. **dish_cv** wird freigegeben
    11. Essenszeit abwarten
    12. **dish_cv** wird wieder gelockt
    13. Essplatz wird wieder freigegeben (**dish_eating**)
    14. Anzahl essender Tiere wird erniedrigt (**num_eating_animals**)
    15. Scheduler wird benarichtigt, dass ein Essplatz verfügbar ist (**dish_cv**)
    16. Falls keine Tiere mehr essen, wird der Scheduler benarichtigt. 
        * Scheduler wartet möglicherweise auf Grund der "Starvation Border" auf leere Futterstellen, um die Tierart zu wechseln(genauer s. 2.5.3)
    17. **dish_cv** wird freigegeben

    
* zum Schutz der Wartelisten wird jeweils ein Mutex verwaltet, der beim Zugriff der Listen gelockt und dann wieder freigegeben wird

### 2.5.3 Funktion scheduler()
Das allgemeine Schedulerverfahren wird in *2.1* beschrieben.
Diese Funktion beschreibt das genaue Vorgehen des Schedulers und beherbergt somit die Haupt-Logik. 
* Der Scheduler besitzt zwei Variablen:
    * **animaltype:** Der Tier-Typ, der gerade die Berechtigung zu essen hat
    * **count:** Zähler um "die Verhungerungsgrenze" einzuhalten
* Die folgende Schleife wird solange durchlaufen bis das Flag, nachdem die Tierthreads beendet sind, gesetzt wird(s. 2.2 Globale Variablen: **flag_finished**)
    * Wenn es keine freien Essplätze gibt, wartet der Scheduler auf das Signal eines Tierthreads, der die Futterstelle verlässt(**dish_cv**)
    * Wenn sich Tiere in der Wartschlange (*s. 2.2 Globale Variablen:* **animals**) befinden, gibt es drei Möglichkeiten wie der Scheduler vorgeht:
        1. Alle Futterstellen sind frei
            * das erste Tier aus der Warteschlange **animals** wird benarichtigt
                * dazu weckt er das erste Tier aus der spezifischen Tierliste auf (*s. 2.2 Globale Variablen:* **cats/dogs/mice**)
            * **animaltype** wird auf den Tiertyp des ausgewählten Tieres gesetzt
            * count wird auf 0 gesetzt
        2. Die Verhungerungsgrenze des ersten Tieres in der Warteschlange ist erreicht: **count** > 3
            * wenn es noch essende Tiere gibt, wartet der Scheduler bis das Signal(2.4.1: 16.), dass es keine essende Tiere mehr gibt, von einem Tierthread kommt.
                * das ist nötig um die Tierart **animaltype** zu wechseln, sodass das erste Tier aus **animals** essen kann -> Bedingung 1
        3. wenn Bedingung 1 und 2 nicht zutrifft
            * der Scheduler durchsucht **animals** auf die Tiertypen, der drei am längsten wartenden Tiere in der Warteliste(d.h. die ersten drei)
                * Wenn einer der Tiertypen, mit dem gerade essendem Tiertyp **animaltyp** übereinstimmt:
                    * **count** wird auf 0 gesetzt, wenn es das am längsten wartende Tier ist(am Anfang der Liste), sonst wird **count** erhöht
                    * das ausgewälte Tier wird aufgeweckt (aus spezifischen Tierliste: **cats/dogs/mice**)
            * bei dem Fall, dass kein passender Tiertyp gefunden worden ist, geht es wieder zum Anfang der Schleife
                * es kommt möglicherweiße zu einem Tiertypwechsel(1.)

                
* zum Schutz der Warteliste **animals** wird natürlich ein Mutex verwaltet, der beim Zugriff gelockt und  wieder freigegeben wird

### 3.1.1 Messungen der Wartezeiten mit Standartwerten
**Parameter:**
- Anzahl Katzen: 6 (Default)
- Anzahl Hunde: 4 (Default)
- Anzahl Mäuse: 2 (Default)
- Anzahl Mahlzeiten: 5 (Default, je Tier)

Die erste Messung gibt einen kurzen Einblick in die Werte des Schedulerst, indem er mit den Standard-Werten ausgeführt wird. Im Diagramm werden die minimalen, durchschnittlichen und Maximalen Wartezeiten angezeigt.

#### Ergebnis
Die durchschnittliche Wartezeiten der Tierarten sind fast gleich, was sich auch in den folgenden Messungen bestätigen wird. Nur das Minimum, besonders von den Mäusen, unterscheidet sich, da sie in diesem Fall zuerst an den Näpfen waren. Das Maximum bei den Mäusen ist auch das höchste, bei dieser Art gibt es also eine große Varianz.

````python
%matplotlib inline
import matplotlib.pyplot as plt
import os
import numpy as np

list_max = []
list_min = []
list_avg = []

def addtolists(lines):
    sl = lines[i+1].split(' ')[1].rstrip('\n')
    list_min.append(float(sl))
    sl = lines[i+2].split(' ')[1].rstrip('\n')
    list_max.append(float(sl))
    sl = lines[i+3].split(' ')[1].rstrip('\n')
    list_avg.append(float(sl))

with open("default", 'r+') as f:
    lines = f.readlines()
    for i in range(0, len(lines)):
        line = lines[i]
        if line.startswith("Cats") or line.startswith("Dogs") or line.startswith("Mice"):            
            addtolists(lines)

index = np.arange(3)
width = 0.30

fig, ax = plt.subplots()

rect1 = plt.bar(index, list_min, width, color='r', label='Minimun')
rect2 = plt.bar(index + width , list_max, width, color='b', label='Maximum')
rect3 = plt.bar(index + width+width, list_avg, width, color='g', label='Durchschnitt')

plt.xticks(index + width + width/2, ('Katzen', 'Hunde', 'Mäuse'))
plt.legend()
plt.ylabel('Zeit in s')
plt.title('Durchschnittliche Wartezeit mit Default-Werten')
axes = plt.gca()
axes.set_ylim([0,9])

plt.show()
````

### 3.1.2 Futterstellen erhöhen
**Parameter:**
- Zahl Tiere: je 5
- Anzahl Mahlzeiten: je 10
- Sättigungszeit: Default der jeweiligen Art
- Anzahl der Futterstellen: wird erhöht

Dieser Test beschreibt, wie sich der Scheduler mit geringen Anzahlen von
Futterstellen, die kleiner sind als die Anzahl der Tier sind, verhält.
Hier kommt es darauf an, wie hoch die Wartezeiten der Tierarten sind,
wenn das Futter begrenzt ist.

Die Parameter, außer die Anzal der Futterstellen, bleiben bei jedem
Durchlauf gleich, entsprechen jedoch nicht den Default-Werten. Die Zahl
der Tiere pro Art wird auf je 5 gesetzt. Die Anzahl der Mahlzeiten der
einzelnen Tiere wird auf 10 erhöht um die Luafzeit zu erhöhen und die
Aussage verdeutlicht wird. Die Sättigungszeit der Tierarteb bleibt
gleich, weil sich sonst alle Tiere gleich verhalten würden.

#### Ergebnis

Wie zu erwarten war sinkt die Durchschnittliche Wartezeit der Tiere mit
einer zunehmenden Anzahl von Futterstellen. Weil mehr Tiere gleichzeitig
essen können, wird mehr Durchsatz möglich und als Resultat sinkt die
Wartezeit.

Daran, dass die Wartezeit aller Tiere sinkt, sieht man, dass keine Art
benachteiligt wird.

Dass die Wartezeit von 3 auf 4 Futterstellen nich wesentlich abnimmt,
liegt eventuell an einer zwischenzeitlich anderen Last auf dem System
oder an einer ungünstigen initialen Zuteilung des Schedulers; trotzdem
ist der Trend eindeutig.

Ab 5 Futterstellen kann die Durchshnittliche Wartezeit nicht weiter,
sinken da nicht mehr als die 5 Tiere einer Art gleichzeitig an die
Futterstelle dürfen. Das spiegelt auch unser Ergebnis mit 6 Futternäpfen
wieder.

````python
avg_cats = []
avg_dogs = []
avg_mice = []

for name in ['dish_1', 'dish_2', 'dish_3', 'dish_4', 'dish_5', 'dish_6']:
    with open(name, 'r+') as f:
        lines = f.readlines()
        for i in range(0, len(lines)):
            line = lines[i]
            if line.startswith("Cats"):
                sl = lines[i+3].split(' ')[1].rstrip('\n')
                avg_cats.append(float(sl))
            elif line.startswith("Dogs"):
                sl = lines[i+3].split(' ')[1].rstrip('\n')
                avg_dogs.append(float(sl))
            elif line.startswith("Mice"):  
                sl = lines[i+3].split(' ')[1].rstrip('\n')
                avg_mice.append(float(sl))

plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)
X = np.arange(1,7)

plt.plot(X, avg_cats, color="red", linewidth=2.0, linestyle="-", label='Katzen')
plt.plot(X, avg_dogs, color="blue", linewidth=2.0, linestyle="-", label='Hunde')
plt.plot(X, avg_mice, color="yellow", linewidth=2.0, linestyle="-", label='Mäuse')
plt.xlim(1, 6)
plt.legend(loc='upper right')
plt.ylabel('Durchschnittliche Wartezeit in s')
plt.xlabel('Anzahl Futternäpfe')

plt.show()
````

### 3.1.3 Zahl der Tiere einer Art wird erhöht
**Parameter:**
- Anzahl Mahlzeiten: je 10
- Anzahl Futterstellen: 2 (Default)
- Anzahl Mäuse: 0
- Anzahl Hunde: 5

In unserem dritten Szenario setzen wir die Parameter fest und verändern
die Anzahl der Tiere einer Art, in diesem Fall Katzen - die beliebtesten
Tiere im Internet. Es soll die Auswirkung auf die anderen Tierarten
beobachtet werden, wenn die Größe einer Tierart zunimmt.

Die Anzahl der Mahlzeiten pro Tier wird wie im vorherigen Szenario im
Vergleich zu den Default-Werten verdoppelt, um die Tiere länger
beobachten zu können. Für die Futterstellen wird der Default-Wert 2
gewählt. Diesmal gibt es keine Mäuse und 5 Hunde. Es wird also
dargestellt, wie Hunde mit einer größer werdenden Anzahl von Katzen
wechselwirken.

#### Ergebnis
In den unten stehenden Grafiken erkennt man, dass mit zunehmender Anzahl der Katzen die durchschnittliche Wartezeit von sowohl Hunden und  Katzen erhöht. Die durchschnittlichen Wartezeiten der Arten unterscheidet sich um den Bruchteil einer Sekunde, was einen einstelligen Prozentwert ausmacht. Dadurch sieht man, dass eine viel größere Population einer Tierart die anderen nicht (wesentlich) benachteiligt.

Auch verhungern keine Tiere, wie man an der maximalen Wartezeiten sehen kann, denn die Maximale Wartezeit beträgt nie die gesamte Laufzeit des Programms. Die Stufen kommen zustande, weil die neu hinzu kommenden Katzen einen neuen Slot belegen, die anderen Tiere also länger warten müssen. Kommt nochmal eine Katze dazu, gesellt sie sich an die noch freie Futterstelle im neuen Slot und braucht so keinen eigenen neuen Slot.

````python
%matplotlib inline
import matplotlib.pyplot as plt
import os
import numpy as np

avg_cats = []
avg_dogs = []
max_cats = []
max_dogs = []

for name in ['catz_0', 'catz_2', 'catz_4', 'catz_6', 'catz_8', 'catz_10', 'catz_12', 'catz_14', 'catz_16', 'catz_18', 'catz_20']:
    with open(name, 'r+') as f:
        lines = f.readlines()
        for i in range(0, len(lines)):
            line = lines[i]
            if line.startswith("Cats"):
                sl = lines[i+2].split(' ')[1].rstrip('\n')
                max_cats.append(float(sl))
                sl = lines[i+3].split(' ')[1].rstrip('\n')
                avg_cats.append(float(sl))
            elif line.startswith("Dogs"):
                sl = lines[i+2].split(' ')[1].rstrip('\n')
                max_dogs.append(float(sl))
                sl = lines[i+3].split(' ')[1].rstrip('\n')
                avg_dogs.append(float(sl))

plt.figure(figsize=(12, 6), dpi=80)
plt.subplot(1, 2, 2)
X = np.arange(0,22,2)

plt.plot(X, avg_cats, color="red", linewidth=2.0, linestyle="-", label='Katzen')
plt.plot(X, avg_dogs, color="blue", linewidth=2.0, linestyle="-", label='Hunde')
plt.xlim(1, 20)
plt.legend(loc='upper right')
plt.ylabel('Durchschnittliche Wartezeit in s')
plt.xlabel('Anzahl der Katzen')


plt.subplot(1, 2, 1)

plt.plot(X, max_cats, color="red", linewidth=2.0, linestyle="-", label='Katzen')
plt.plot(X, max_dogs, color="blue", linewidth=2.0, linestyle="-", label='Hunde')
plt.xlim(1, 20)
plt.legend(loc='upper right')
plt.ylabel('Maximale Wartezeit in s')
plt.xlabel('Anzahl der Katzen')


plt.show()
````

### 3.1.4 Die Ess-Zeit wird variiert
**Parameter:**
- Anzahl Katzen: 6 (Default)
- Anzahl Hunde: 4 (Default)
- Anzahl Mäuse: 2 (Default)
- Anzahl Futterstellen: 2 (Default)
- Anzahl Mahlzeiten: 5 (alle Tiere, Default)
- Untere Grenze der Fresszeit: 0.5

In diesem Szenario geht es darum, die Ess-Zeit - die der Rechenzeit
eines Prozesses entspricht - zu variieren. Dadurch wird das Verhalten
des Schedulers auch "zufälliger" und entspricht mehr dem Verhalten in
einem echten System, da es dort auch Prozesse gibt, die verschieden
lange auf der CPU rechnen.

Für diesen Test bleiben alle Werte bei ihrem jeweiligen Standard; nur
die untere und obere Ranndom-Grenze für das Essen werden verändert.
Die untere Grenze wir auf 0.5 festgesetzt, was der Hälfte des
Original-Werts entspricht. Die obere Grenze wird stückweise erhöht.


#### Ergebnis
Wie man am Diagramm sieht, zeigen wir hier nicht nur die
durchschnittliche, sondern auch die maximale Wartezeit. Das soll
verdeutlichen, dass die Tierarten zwar unterschiedlich behandelt
werden, aber nicht verhungern, da der Scheduler skaliert und beendet.

Aus den Diagrammen sieht man, dass die Tierarten unterschiedliche
Wartezeiten haben, die mit der (randomisierten) Ausführungszeit linear
skalieren.

Katzen und Hunde haben dem ersten Diagramm nach die gleichen maximalen
Wartezeiten; die Mäuse sind etwas abgeschlagen darüber. Das liegt wohl
daran, dass die Tierarten grundsätzlich geleich vom Scheduler behandelt
werden, aber die Mäuse schneller wieder Hungrig werden. Hier wird nicht
je nach Ziel des Schedulers nicht genug auf die Eigenheiten der Threads
eingegangen.

Im zweiten Diagramm mit den duchschnittlichen Wartezeiten sieht man,
dass die Mäuse besser dastehen und die Hunde die schlechtesten
durchschnittlichen Wartezeiten haben. Die Katzen haben nach wie vor
ähnliche Werte.

Die Mäuse haben zwischen maximaler und durchschnittlicher Wartezeit eine
Abweichung von etwa 20%. Diese werden also relativ benachteiligt. Die
anderen Arten haben nur eine Abweichung im einstelligen Prozentbereich.

````python
avg_cats = []
max_cats = []
avg_dogs = []
max_dogs = []
avg_mice = []
max_mice = []

for name in ['E1', 'E2', 'E3', 'E4', 'E5', 'E6']:
    with open(name, 'r+') as f:
        lines = f.readlines()
        for i in range(0, len(lines)):
            line = lines[i]
            if line.startswith("Cats"):
                sl = lines[i+2].split(' ')[1].rstrip('\n')
                max_cats.append(float(sl))
                sl = lines[i+3].split(' ')[1].rstrip('\n')
                avg_cats.append(float(sl))
            elif line.startswith("Dogs"):
                sl = lines[i+2].split(' ')[1].rstrip('\n')
                max_dogs.append(float(sl))
                sl = lines[i+3].split(' ')[1].rstrip('\n')
                avg_dogs.append(float(sl))
            elif line.startswith("Mice"):  
                sl = lines[i+2].split(' ')[1].rstrip('\n')
                max_mice.append(float(sl))
                sl = lines[i+3].split(' ')[1].rstrip('\n')
                avg_mice.append(float(sl))

plt.figure(figsize=(12, 6), dpi=80)
plt.subplot(1, 2, 1)
X = np.arange(1,7)

plt.plot(X, max_cats, color="red", linewidth=2.0, linestyle="-", label='Katzen')
plt.plot(X, max_dogs, color="blue", linewidth=2.0, linestyle="-", label='Hunde')
plt.plot(X, max_mice, color="yellow", linewidth=2.0, linestyle="-", label='Mäuse')
plt.xlim(1, 6)
plt.legend(loc='upper right')
plt.ylabel('Maximale Wartezeit in s')
plt.xlabel('Obere Random-Grenze (in s)')

plt.subplot(1, 2, 2)
plt.plot(X, avg_cats, color="red", linewidth=2.0, linestyle="-", label='Katzen')
plt.plot(X, avg_dogs, color="blue", linewidth=2.0, linestyle="-", label='Hunde')
plt.plot(X, avg_mice, color="yellow", linewidth=2.0, linestyle="-", label='Mäuse')
plt.xlim(1, 6)
plt.legend(loc='upper right')
plt.ylabel('Durchschnittliche Wartezeit in s')
plt.xlabel('Obere Random-Grenze (in s)')

plt.show()
````

## 4.1 Kein Verhungern
Das dritte Szenario beschreibt das Verhalten von einer
Tierart, wenn die Population einer anderen Art zunimmt.
Dabei sollte festgestellt werden, ob die bisherige Tierart
von der anderen übergangen wird, wenn deren Anzahl zunnimmt.

Am Diagramm sieht man die Wartezeiten der Arten und merkt,
dass sich die Wartezeiten der beiden beteiligten Arten nur
im einstelligen Prozentbereich unterscheiden. Das Ergebnis
zeigt, dass die Arten gleich behandelt werden und nicht eine
die Wartezeiten der anderen in die Höhe treibt.

Auch im vierten Szenario sieht man, dass mit der zunahme der
Futter-Zeit der Tiere nicht bei einer Art die Wartezeit einer
Art überproportinal in die Höhe geht. Leider warten die Mäuse,
die öfters Wartezeit brauchen länger; das sollte man bei einem
Scheduler für ein interaktives System verbessern, weil
interaktive Prozesse auch wenig rechnen und schnell wieder
Rechnen wollen, also eine kurze Wartezeit brauchen. Sonst wird
die Antwortzeit nicht mehr hinnehmbar.

## 4.2 Gute Auslastung der Futterstellen
Im zweiten Szenario gibt es zu Beginn viel mehr Tiere als
Futterstellen; ab dann wird die Zahl der Futterstellen erhöht.
Mit einer als Basis sinkt die Wartezeit beim erhöhen auf zwei
Futterstellen auf gut die hälfte; bei einer verdreifachung
beträgt sie gut ein drittel. Ein gewisser Overhead ist also zu
verzeichnen, der generelle Trend ist dennoch gut.

Wählt man die Anzahl der Futterstellen größer als die Anzahl der
Tiere, wird die Wartezeit erwartungsgemäß nicht kleiner. Die
Wartezeit wird auch nicht größer, der Scheduler kommt also auch
nicht aus dem Tritt, wenn man die Resourcen nicht auslasten kann.
